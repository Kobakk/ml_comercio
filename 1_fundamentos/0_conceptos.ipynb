## Las variables

### Cadenas

La definición de las cadenas en Python se realiza a través de las comillas simples (') o de comillas dobles ("), pero no mezclándolas. Por convención siempre se suelen utilizar las simples, excepto cuando dentro de la cadena haya que incluir una comilla doble.

```python
cadena1 = 'Que calor hace'
cadena2 = "Que calor hace"
cadena3 = "Que calo' hace"

print(cadena1)
print(type(cadena1))
print(cadena2)
print(type(cadena2))
print(cadena3)
print(type(cadena3))
```

```python
#Otras operaciones muy útiles son las siguientes
some_string = 'Hello World'
print('Operaciones basicas')
print('some_string', some_string)

#comprobar si comienza por ...
print("some_string.startswith('H')", some_string.startswith('H'))
print("some_string.startswith('h')", some_string.startswith('h'))

#comprobar si termina por ...
print("some_string.endswith('d')", some_string.endswith('d'))

#comprobar todas las palabras comienzan por mayúscula y si el resto de letras son minúscula
print('some_string.istitle()', some_string.istitle())

#comprobar si son mayúsculas, minúsculas
print('some_string.isupper()', some_string.isupper())
print('some_string.islower()', some_string.islower())

#comprobar si todos los caracteres son letras
print('some_string.isalpha()', some_string.isalpha())
print('')

#también podemos realizar conversiones
print('Conversiones')
print('some_string.upper()', some_string.upper()) #mayúsculas todo
print('some_string.lower()', some_string.lower()) #minúsculas todo
print('some_string.title()', some_string.title()) #cada palabra comienza en mayúscula y el resto son minúsculas
print('some_string.swapcase()', some_string.swapcase()) #intercambiar mayúsculas por minúsculas y viceversa
print('String leading, trailing spaces', " xyz ".strip()) #eliminar espacios al principio y final
```

##### Operaciones aritméticas

```python
home = 10
away = 15

#suma
print(home + away)
print(type(home + away))

#multiplicación
print(10 * 4)
print(type(10*4))

#resta
goals_for = 10
goals_against = 7
print(goals_for - goals_against)
print(type(goals_for - goals_against))

#división con decimales
print(100 / 21)
print(type(100 / 20))

res1 = 3/2
print(res1)
print(type(res1))

#división entera
res1 = 3//2
print(res1)
print(type(res1))

#modulo
print('Modulus division 4 % 2:', 4 % 2)
print('Modulus division 3 % 2:', 3 % 2)

#exponencial
a = 5
b = 3
print(a ** b)
```

### Booleanos

Las variables booleanas únicamente pueden ser del tipo *True* o *False*. El equivalente de int y float en los booleanos es bool.


```python
all_ok = True
print(all_ok)
all_ok = False
print(all_ok)
print(type(all_ok))
```

Se puede traducir el valor booleano a entero:

```python
print(int(True))
print(int(False))
print(bool(1))
print(bool(0))
```

##### Operadores lógicos


```python
print(True and True)  #operador AND
print(True or False)  #operador OR
print(not False)      #operador NOT
```

### Condicional **if**

```python
num = int(input('Enter a number: '))
if num < 0:
  print(num, 'is negative')
```

### Condicional **if ... else ...**

```python
num = int(input('Enter yet another number: '))
if num < 0:
  print('Its negative')
else:
  print('Its not negative')
```

### Condicional **if ... else if ... else ...**

```python
savings = float(input("Enter how much you have in savings: "))
if savings == 0:
  print("Sorry no savings")
elif savings < 500:
  print('Well done')
elif savings < 1000:
  print('Thats a tidy sum')
elif savings < 10000:
  print('Welcome Sir!')
else:
  print('Thank you')
```

### Bucle con **while**

```python
count = 0
print('Starting')
while count < 10:
  print(count, ' ', end='') # part of the while loop
  count += 1    # also part of the while loop
print()
print('Done')
```

### Bucle con **for**

```python
#contador ascendente
print('Print out values in a range')
for i in range(0, 10):
  print(i, ' ', end='') # part of the while loop
print()
print('Done')

#contador descendente
print('Print out values in a range')
for i in range(10, 0, -1):
  print(i, ' ', end='') # part of the while loop
print()
print('Done')
```

## Definición de funciones

```python
#sin parametros de entrada
def print_msg():
  print('Hola mundo')

print_msg()

#con parametros de entrada
def print_my_msg(msg):
  print(f'Mensaje: {msg}')

print_my_msg('Hola mundo')

#con devolución de un valor
def cuadrado(n):
  return n**2

print(cuadrado(3))

#con devolución de varios valores
def invertir(a, b):
  return b,a

print(invertir(1,2))

#varios parametros de entrada
def greeter1(name, message):
  print(f'Welcome {name} - {message}')

greeter1('Carlos', 'Bienvenido')

#parámetros con valor por defecto
def greeter2(name, message = 'Live Long and Prosper'):
  print(f'Welcome {name} - {message}')

greeter2('Carlos')

#cualquier numero de parametros
def greeter3(*args):
  for name in args:
    print('Welcome', name)

greeter3('Paula', 'Ada', 'Carlota')

```
