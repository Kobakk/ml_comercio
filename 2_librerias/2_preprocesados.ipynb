# Generación de datasets de prueba

```python
import numpy as np

# Si usamos una semilla, los datos aleatorios serán reproducibles
np.random.seed(30)
dimension=10


edades=np.random.randint(18,36,dimension)
alturas=np.random.uniform(1.50,1.90,dimension)
pesos=np.random.randint(60,101,dimension)
# type(edades)
# print(edades)
# print(alturas)
# print(pesos)
muestras = np.array([edades,alturas,pesos])
muestras =muestras.T

# Generar muestras sin transposicion

muestras2= np.column_stack([edades,alturas,pesos])
print(muestras2)
```

# Crear una funcion que admita como parámetro un array. La mostrará la media, mediana, moda,desviacion, min y max, percentiles




```python
import numpy as np

def resumen_estadistico(arr):
    arr = np.array(arr)

    # MEDIA
    media = np.mean(arr)

    # MEDIANA
    mediana = np.median(arr)

    # MODA
    valores, conteos = np.unique(arr, return_counts=True)
    moda = valores[conteos.argmax()]

    # DESVIACIÓN ESTÁNDAR
    desviacion = np.std(arr)

    # MIN y MAX
    minimo = np.min(arr)
    maximo = np.max(arr)

    # PERCENTILES (25, 50, 75)
    p25 = np.percentile(arr, 25)
    p50 = np.percentile(arr, 50)
    p75 = np.percentile(arr, 75)

    print("Media:", media)
    print("Mediana:", mediana)
    print("Moda:", moda)
    print("Desviación estándar:", desviacion)
    print("Mínimo:", minimo)
    print("Máximo:", maximo)
    print("Percentil 25:", p25)
    print("Percentil 50:", p50)
    print("Percentil 75:", p75)
```

```python
columnas = ["edad", "altura", "peso"]

for i, nombre in enumerate(columnas):
    col = muestras2[:, i]  # slice de la columna i
    print(f"---------{nombre}---------")
    resumen_estadistico(col)

```

# Crear una función que pasado un array devuelva un array con los posibles outliers

```python
def detecta_outliers(arr):
  arr= np.array(arr)
  Q1 = np.percentile(arr,25)
  Q3 = np.percentile(arr,75)
  IQR = Q3-Q1
  lim_inf= Q1-IQR*1.5
  lim_sup= Q3+IQR*1.5
  return  arr[ (arr < lim_inf) | (arr>lim_sup)]

```

```python


columnas = ["edad", "altura", "peso"]

for i, nombre in enumerate(columnas):
    col = muestras2[:, i]  # slice de la columna i
    print(f"---------{nombre}---------")
    print(detecta_outliers(col))

#alteramos algún valor
muestras2[ 2, 1]=16.15
print(muestras2)

for i, nombre in enumerate(columnas):
    col = muestras2[:, i]  # slice de la columna i
    print(f"---------{nombre}---------")
    print(detecta_outliers(col))
```

# Crear una función que normalize a 0 y 1 un array que se le pasa como parámetro. Devolverá un nuevo array normalizado.

### Fórmula de normalización Min-Max

1. **Identificar los valores mínimo y máximo del array**

   * Se busca el menor valor ((x_min)) y el mayor valor ((x_max})) del conjunto de datos.
   * Esto permite establecer el rango original de los datos.

2. **Restar el valor mínimo a cada elemento**

   * Se hace (x_i - x_min) para cada elemento (x_i).
   * Esto desplaza todos los valores para que el mínimo del array pase a ser 0.

3. **Dividir entre el rango (máximo - mínimo)**

   * Se divide lo obtenido en el paso anterior entre (x_max - x_min).
   * Esto escala los valores proporcionalmente para que el máximo del array se convierta en 1.

4. **Caso especial: todos los valores iguales**

   * Si (x_max = x_min), el rango es 0 y la división no está definida.
   * En este caso, se devuelve un array de ceros (o cualquier valor constante), porque todos los elementos son iguales y ya están “normalizados”.

5. **Resultado final**

   * Se obtiene un array donde el valor mínimo original es 0, el máximo es 1 y todos los demás valores están proporcionales entre 0 y 1.
   * Esta transformación es útil para **algoritmos de Machine Learning, comparaciones entre datos o visualizaciones**, donde es importante tener todos los valores en la misma escala.



```python
import numpy as np

def normalizar_array_vectorizado(arr):
    arr = np.array(arr, dtype=float)
    min_val = arr.min()
    max_val = arr.max()

    # Evitamos división por cero si todos los valores son iguales
    return (arr - min_val) / (max_val - min_val) if max_val != min_val else np.zeros_like(arr)
```

```python
columnas = ["edad", "altura", "peso"]

for i, nombre in enumerate(columnas):
    col = muestras2[:, i]  # slice de la columna i
    print(f"---------{nombre}---------")
    print(normalizar_array_vectorizado(col))
```
